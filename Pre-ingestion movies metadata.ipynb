```python
import pandas as pd
import json
import ast
import numpy as np
```

```python
def json_to_df(df, column):
    list = []

    for entry in df[column]:
        if pd.notna(entry):
            try:
                data = ast.literal_eval(entry)
                list.append(data)
            except (SyntaxError, ValueError):
                list.append([])
        else:
            list.append([])

    dummy_df = pd.DataFrame()
    for i, item in enumerate(list):
        df = pd.DataFrame([list[i]])
        df["index"] = i
        dummy_df = pd.concat([dummy_df, df], axis=0, ignore_index=True)

    return dummy_df
```

### Important note: 
The file consists of four columns containing JSON objects. However, the JSON is not properly formatted. Therefore it is required to perform transformations to enable unpacking the data and loading it into database.

### 1. Parsing "belongs_to_collection" column

```python
belongs_to_collection_filtered = movies_metadata[movies_metadata['belongs_to_collection'].notna()].reset_index()
belongs_to_collection_filtered
```

```python
collection_list = []

for entry in belongs_to_collection_filtered['belongs_to_collection']:
    if pd.notna(entry):
        try:
            collection_data = ast.literal_eval(entry)
            collection_list.append(collection_data)
        except (SyntaxError, ValueError):
            collection_list.append([])
    else:
        collection_list.append([])
```

We received a list with cast, but still it is needed to link the data with movie id. For this purpose the index will be used.

```python
collection_df = pd.DataFrame()
for i, item in enumerate(collection_list):
    df = pd.DataFrame([collection_list[i]])
    #df["index"] = i
    collection_df = pd.concat([collection_df, df], axis=0, ignore_index=True)

collection_df.head()
```

```python
collection_df = pd.concat([collection_df, belongs_to_collection_filtered['index']], axis=1)
collection_df
```

Now, as all required transformations are done, the data will be saved to .csv format for two reasons. First, to be reused later on and the second not to loose the effects, since those transformations were quite time consuming. 

```python
collection_df.to_csv('data/ready_for_ingestion/collections_stage.csv', index=False)
```

### 2. Parsing "genres" column

```python
list = []

for entry in movies_metadata["genres"]:
    if pd.notna(entry):
        try:
            data = ast.literal_eval(entry)
            list.append(data)
        except (SyntaxError, ValueError):
            list.append([])
    else:
        list.append([])
        
list
```

```python
dummy_df = pd.DataFrame()
for i, item in enumerate(list):
    df = pd.DataFrame([list[i]])
    df["index"] = i
    dummy_df = pd.concat([dummy_df, df], axis=0, ignore_index=True)
```

```python
dummy_df
```

```python
another_dummy_df = pd.DataFrame()
genres_list = dummy_df.values.tolist()
for i, item in enumerate(genres_list):
    for j in genres_list[i]:
        df = pd.DataFrame([j])
        df["index"] = i
        another_dummy_df = pd.concat([another_dummy_df, df], axis=0, ignore_index=True)
```
